```python
# EDA - Home Office Sponsors Register
# checking the list of licensed sponsors - this is training data for my classifier later

import pandas as pd
import matplotlib.pyplot as plt

# loading the sponsors file
df = pd.read_csv('HO_Sponsors_13Jun2026.csv')

# basic checks
print("shape:", df.shape)
print("\ncolumns:", df.columns.tolist())
print("\nfirst 5 rows:")
print(df.head())
print("\nmissing values:")
print(df.isnull().sum())
```

```python
# checking the different routes - skilled worker is the one i care about
print("routes available:")
print(df['Route'].value_counts())
print("\nskilled worker count:", len(df[df['Route'] == 'Skilled Worker']))
```

```python
# top 20 cities with most sponsors
top_cities = df['Town/City'].value_counts().head(20)
print("top 20 cities:")
print(top_cities)
```

```python
# plotting top 20 cities
fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('#f8fafc')
ax.set_facecolor('#f8fafc')

top_cities.sort_values().plot(kind='barh', ax=ax, color='#2563eb')

ax.set_title('Top 20 Cities by Number of Sponsors', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Number of Sponsors', fontsize=11)
ax.set_ylabel('City', fontsize=11)
ax.spines[['top', 'right']].set_visible(False)
ax.grid(axis='x', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.savefig('sponsors_top_cities.png', dpi=150, bbox_inches='tight')
plt.show()
print("done!")
```
