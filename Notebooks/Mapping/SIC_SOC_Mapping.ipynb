```python
# SIC/SOC 2020 mapping table
# mapping all datasets to my 5 sectors so i can merge them later

import pandas as pd

# my 5 sectors and which SIC codes belong to each
# SIC codes are what NOMIS uses

sic_mapping = {
    'Technology': [62, 63, 58, 61],        # software, IT services, telecoms
    'Healthcare': [86, 87, 88],             # hospitals, residential care, social work
    'Finance': [64, 65, 66],               # banking, insurance, financial services
    'Engineering': [71, 72, 33, 28, 25],   # engineering consultancy, R&D, manufacturing
    'Education': [85]                       # education
}

# turning it into a dataframe
rows = []
for sector, codes in sic_mapping.items():
    for code in codes:
        rows.append({'SIC_Code': code, 'Sector': sector})

df_sic = pd.DataFrame(rows)
print("SIC mapping table:")
print(df_sic)
print("\ntotal SIC codes mapped:", len(df_sic))
```

```python
# SOC 2020 codes - what ASHE and ImmigStats use
# these are occupation codes not industry codes

soc_mapping = {
    'Technology': [21, 213, 2135, 2136, 2137, 2139],      # IT professionals
    'Healthcare': [22, 223, 224, 225, 226],                 # health professionals
    'Finance': [24, 241, 242, 243],                         # finance and business professionals
    'Engineering': [21, 211, 212, 214, 215],               # science and engineering professionals
    'Education': [23, 231, 232, 233]                        # teaching professionals
}

rows = []
for sector, codes in soc_mapping.items():
    for code in codes:
        rows.append({'SOC_Code': code, 'Sector': sector})

df_soc = pd.DataFrame(rows)
print("SOC 2020 mapping table:")
print(df_soc)
print("\ntotal SOC codes mapped:", len(df_soc))
```

```python
# ImmigStats uses industry name strings not codes
# so mapping those too

industry_mapping = {
    'Technology': ['Information and Communication', 'Information and Communications'],
    'Healthcare': ['Human Health and Social Work Activities'],
    'Finance': ['Financial and Insurance Activities'],
    'Engineering': ['Professional, Scientific and Technical Activities',
                    'Manufacturing'],
    'Education': ['Education']
}

rows = []
for sector, industries in industry_mapping.items():
    for industry in industries:
        rows.append({'Industry_Name': industry, 'Sector': sector})

df_industry = pd.DataFrame(rows)
print("Industry name mapping table:")
print(df_industry)

# saving all 3 mapping tables to csv
df_sic.to_csv('SIC_Mapping.csv', index=False)
df_soc.to_csv('SOC_Mapping.csv', index=False)
df_industry.to_csv('Industry_Mapping.csv', index=False)
print("\nall 3 mapping tables saved!")
```

```python
# quick summary of the mapping
print("mapping table summary:")
print(f"- SIC codes mapped: {len(df_sic)} codes across 5 sectors")
print(f"- SOC 2020 codes mapped: {len(df_soc)} codes across 5 sectors")
print(f"- Industry name strings mapped: {len(df_industry)} strings across 5 sectors")
print("\nfiles saved:")
print("- SIC_Mapping.csv")
print("- SOC_Mapping.csv")
print("- Industry_Mapping.csv")
print("\nnote: SOC code 21 appears in both Technology and Engineering")
print("will handle this overlap during cleaning by checking description text")
```
