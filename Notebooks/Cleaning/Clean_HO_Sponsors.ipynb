```python
# Data Cleaning - Home Office Sponsors Register
# cleaning and standardising the sponsors list ready for the classifier

import pandas as pd

# loading the raw file
df = pd.read_csv('HO_Sponsors_13Jun2026.csv')

print("shape:", df.shape)
print("\ncolumns:", df.columns.tolist())
print("\nfirst 5 rows:")
print(df.head())
```

```python
# keeping only skilled worker route - thats what matters for my platform
df_clean = df[df['Route'] == 'Skilled Worker'].copy()
print("skilled worker only shape:", df_clean.shape)

# standardising city names - london vs LONDON issue found in EDA
df_clean['Town/City'] = df_clean['Town/City'].str.strip().str.title()

# checking london is fixed
london_count = df_clean[df_clean['Town/City'] == 'London'].shape[0]
print("london count after standardising:", london_count)

# dropping county column - 94039 missing values, not needed
df_clean = df_clean.drop(columns=['County'])

# renaming columns to simpler names
df_clean.columns = ['Organisation', 'City', 'Type_Rating', 'Route']

print("\nshape after cleaning:", df_clean.shape)
print("\nmissing values:")
print(df_clean.isnull().sum())
```

```python
# checking top cities after standardising
print("top 10 cities after standardising:")
print(df_clean['City'].value_counts().head(10))
```

```python
# saving clean file
df_clean.to_csv('HO_Sponsors_Clean.csv', index=False)
print("saved! shape:", df_clean.shape)
print("\nfirst 5 rows:")
print(df_clean.head())
```
